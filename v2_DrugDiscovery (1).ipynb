Step 2: Convert SMILES to Molecule

```python
def smiles_to_molecule(smiles):
    """Convert a SMILES string to an RDKit molecule object."""
    molecule = Chem.MolFromSmiles(smiles)
    if molecule is None:
        print(f"Invalid SMILES: {smiles}")
    return molecule
```

```python
smiles = "COc1ccc2nccc([C@@H](O)CC[C@@H]3CCN(C4CC(c5ccccc5C(F)(F)F)C4)C[C@@H]3C(=O)O)c2c1"  # Example SMILES for ethanol
molecule = smiles_to_molecule(smiles)
print(molecule)  # Should return an RDKit Mol object
```

```python
def get_node_features(molecule):
    """Extract features for each atom in the molecule."""
    node_features = []
    for atom in molecule.GetAtoms():
        features = [
            atom.GetAtomicNum(),         # Atomic number
            atom.GetFormalCharge(),      # Formal charge
            atom.GetTotalNumHs(),        # Number of hydrogens
            atom.GetHybridization(),     # Hybridization state
            atom.GetIsAromatic()         # Aromaticity
        ]
        node_features.append(features)
    return node_features
```

```python
def get_edge_features(molecule):
    """Extract features for each bond in the molecule."""
    edge_features = []
    for bond in molecule.GetBonds():
        features = [
            bond.GetBondTypeAsDouble(),  # Bond type as a numeric value
            bond.GetIsConjugated(),     # Conjugation
            bond.IsInRing()             # Ring membership
        ]
        edge_features.append(features)
    return edge_features
```

```python
edge_features = get_edge_features(molecule)
print(edge_features)
# Output: List of features for each bond in the molecule
```

```python
import pandas as pd


# Initialize lists for features
node_features_list = []
edge_features_list = []

# Loop through each row in the dataframe
for index, row in df_cleaned.iterrows():
    smiles = row['Smiles']
    molecule = smiles_to_molecule(smiles)

    if molecule:  # If molecule conversion is successful
        node_features = get_node_features(molecule)
        edge_features = get_edge_features(molecule)
    else:
        node_features = None
        edge_features = None

    # Append to the lists
    node_features_list.append(node_features)
    edge_features_list.append(edge_features)

# Add features as new columns in the dataframe
df_cleaned['Node_Features'] = node_features_list
df_cleaned['Edge_Features'] = edge_features_list
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from rdkit import Chem

# Define the unique categories for hybridization and bond types
hybridization_types = ['SP', 'SP2', 'SP3']
bond_types = ['SINGLE', 'DOUBLE', 'TRIPLE', 'AROMATIC']

# Initialize one-hot encoders with sparse_output=False
hybridization_encoder = OneHotEncoder(categories=[hybridization_types], sparse_output=False)
bond_type_encoder = OneHotEncoder(categories=[bond_types], sparse_output=False)

# Fit the encoders
hybridization_encoder.fit(np.array(hybridization_types).reshape(-1, 1))
bond_type_encoder.fit(np.array(bond_types).reshape(-1, 1))

def encode_node_features(node_features):
    encoded_features = []
    for atom in node_features:
        atomic_num, formal_charge, num_hs, hybridization, is_aromatic = atom

        # Encode hybridization
        hybridization_str = hybridization.name
        hybridization_encoded = hybridization_encoder.transform([[hybridization_str]]).flatten()

        # Encode aromaticity
        aromaticity_encoded = [1 if is_aromatic else 0]

        # Combine all features
        combined_features = [atomic_num, formal_charge, num_hs] + list(hybridization_encoded) + aromaticity_encoded
        encoded_features.append(combined_features)
    return encoded_features

def encode_edge_features(edge_features):
    encoded_features = []
    for bond in edge_features:
        bond_type, is_conjugated, is_in_ring = bond

        # Ensure bond_type is a string representation
        if isinstance(bond_type, Chem.rdchem.BondType):
            bond_type_str = bond_type.name
        else:
            # Handle cases where bond_type is a float
            bond_type_mapping = {
                1.0: 'SINGLE',
                2.0: 'DOUBLE',
                3.0: 'TRIPLE',
                1.5: 'AROMATIC'
            }
            bond_type_str = bond_type_mapping.get(bond_type, 'UNKNOWN')

        # Encode bond type
        bond_type_encoded = bond_type_encoder.transform([[bond_type_str]]).flatten()

        # Encode conjugation and ring membership
        conjugation_encoded = [1 if is_conjugated else 0]
        ring_encoded = [1 if is_in_ring else 0]

        # Combine all features
        combined_features = list(bond_type_encoded) + conjugation_encoded + ring_encoded
        encoded_features.append(combined_features)
    return encoded_features

# Assuming df_cleaned is your DataFrame
df_cleaned['Encoded_Node_Features'] = df_cleaned['Node_Features'].apply(encode_node_features)
df_cleaned['Encoded_Edge_Features'] = df_cleaned['Edge_Features'].apply(encode_edge_features)
```
